# Updating Rows and Columns: Modifying Data within our DataFrames

````python
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["CoreyMSchafer@gmail.com", "JaneDoe@email.com", "JohnDoe@email.com"]
}
````

````python
import pandas as pd
````

````python
df = pd.DataFrame(people)
````

````python
df
````

````python
df.columns
````

## Updating Columns

To rename columns, pass in list to columns attribute:

````python
df.columns = ['first name', 'last name', 'email']
````

````python
df
````

Change all columns to uppercase using a list comprehension:

````python
df.columns = [x.upper() for x in df.columns]
````

````python
df
````

Replace spaces with underscores:

````python
df.columns = df.columns.str.replace(' ', '_')
````

````python
df
````

````python
df.columns = [x.lower() for x in df.columns]
````

````python
df
````

Just change some columns, using rename method:

````python
df.rename(columns={'first_name': 'first', 'last_name': 'last'})
````

````python
df
````

But changes with .rename method do not persist! To make permanent, use "inplace=True":

````python
df.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
````

````python
df
````

## Updating Rows

First, let's update a couple of values. Just change John's last name:

Check John's row ([2]):

````python
df.loc[2]
````

Now replace values with a new array:

````python
df.loc[2] = ['John', 'Smith', 'JohnSmith@email.com']
````

````python
df
````

But we can also change it by specifying column as well as row. First read the 'last' and 'email' columns for row 2:

````python
df.loc[2, ['last', 'email']]
````

Now pass in array value for the target columns in row two:

````python
df.loc[2, ['last', 'email']] = ['Doe', 'JohnDoe@email.com']
````

````python
df
````

Now let's change a single value:

````python
df.loc[2, 'last'] = 'Smith'
````

````python
df
````

Or use the 'at' method:

````python
df.at[2, 'last'] = 'Doe'
````

````python
df
````

Here's what happens when you try changing without using the .loc or .at method:

````python
filt = (df['email'] == 'JohnDoe@email.com')
df[filt]
````

````python
filt = (df['email'] == 'JohnDoe@email.com')
df[filt]['last']
````

````python
df[filt]['last'] = 'Smith'
````

````python
df
````

See the explanation in the error message link for details, but the short version is that the way
we're doing this requires multiple operations in the background to determine whether pandas
returns a view or a copy of our dataset. Here, our value is only getting set on a temporary object
that pandas discards immediately after.

````python
filt = (df['email'] == 'JohnDoe@email.com')
df.loc[filt, 'last'] = 'Smith'
````

````python
df
````

### More Advanced Methods
Four more advanced methods to change multiple rows at once:

* apply
* map
* applymap
* replace

#### Apply

First, look at apply. Say we want to check the length of each email value:

````python
df['email'].apply(len)
````

Now change all emails to uppercase using simple function:

````python
def update_email(email):
    return email.upper()
````

````python
df['email'].apply(update_email)
````

````python
df
````

"Apply" the change by redefining all values in the column:

````python
df['email'] = df['email'].apply(update_email)
````

````python
df
````

Here's how to do the same thing with a lambda (anonymous) function:

````python
df['email'] = df['email'].apply(lambda x: x.lower())
````

````python
df
````

The above were all series objects, now we'll look at how it works with dataframes.

Running apply on a dataframe runs whatever function is passed on the entire dataframe.

This will give us the length of each value in the email series:

````python
df['email'].apply(len)
````

Now try to run method on entire dataframe:

````python
df.apply(len)
````

Here the function isn't being applied to every value in the frame, but to each series in the frame:
the column. So what it's telling is the number of rows in each column!

Now let's apply the method to one of the columns:

````python
len(df['email'])
````

You can also have it apply to rows by changing the axis:

````python
df.apply(len, axis='columns')
````

So what we want to do is use a function that will make sense on a series.

````python
df.apply(pd.Series.min)
````

````python
df.apply(lambda x: x.min())
````

This would make more sense on numerical data.

Running apply on a series applies the function to every value in the series.

Running apply on a dataframe applies it to every series in the dataframe.

#### Applymap

To run a function on every value in the dataframe you would use applymap,
which only works on dataframes.

````python
df.applymap(len)
````

Now function is acting on each value in the dataframe.

Here is how to make all values in the dataframe lowercase:

````python
df.applymap(str.lower)
````

#### Map
Map only works on a series (like apply). Used for substituting each value in a series
with a different value.

Here is example of map method being used on a series:

````python
df['first'].map({'Corey': 'Chris', 'Jane': 'Mary'})
````

Values we didn't substitute got converted to NaN!

But what if we wanted to leave rest of values alone?

Then use replace method:

#### Replace

````python
df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
````

Note: Everything done above didn't change the dataframe. To do that, we need to
redefine:

````python
df['first'] = df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
````

````python
df
````

### Map
What is we want to change the responses a true boolean value (True or False)?

Use the map method:

````python
df['Hobbyist'].map({'Yes': True, 'No': False})
````

Since there's no inplace argument for map, we can just set the series to equal the mapping:

````python
df['Hobbyist'] = df['Hobbyist'].map({'Yes': True, 'No': False})
````

````python
df
````

Keep in mind that using map anything outside of a Yes or No answer in the column will be converted to a NaN.

If we only wanted to change the Yes and No answers, we'd use replace. But since this column ('Hobbyist')
only has Yes or No, it's safe to use map.
