# Cleaning Data: Casting Datatypes and Handling Missing Values


```python
import pandas as pd
import numpy as np
```

## Sample Data

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

### Missing Values

Drop NaN values:

```python
df.dropna()
```

Expand to show default parameters for dropna:

```python
df.dropna(axis='index', how='any')
```

Here, 'index' is for rows, 'any' means match any cell.

Changing how to 'all' returns all rows that have at least one cell without NaN (in row 4, ever cell was NaN):

```python
df.dropna(axis='index', how='all')
```

Set axis to 'columns' to filter by columns:

```python
df.dropna(axis='columns', how='all')
```

With this data, combining axis=columns and how=any returns nothing for each column, since every column contains at least one missing value.

```python
df.dropna(axis='columns', how='any')
```

Use 'subset' to filter based on specific columns (rows 3 and 4 contain 'NaN' or 'None' in email column):

```python
df.dropna(axis='index', how='any', subset=['email'])
```

Since every row except 3 and 4 has some non-NaN/None value in e-mail, same result for 'all':

```python
df.dropna(axis='index', how='all', subset=['email'])
```

Filter on two columns, instead of just one, by feeding subset an array:

```python
df.dropna(axis='index', how='all', subset=['last','email'])
```

Need to use "inplace=True" to make permanent change

Replace custom missing values with proper NaN (np.nan) values:

```python
df = pd.DataFrame(people)

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
Repeat previous observations with modified data:
```

```python
df.dropna()
```

```python
df.dropna(axis='index', how='all', subset=['last','email'])
```

Show every cell that has an NaN type value:

```python
df.isna()
```

Replace NaN values with a custom value:

```python
df.fillna('MISSING')
```

Or 0 in numeric fields:

```python
df.fillna(0)
```

### Casting DataTypes

```python
df.dtypes
```

Object datatype is often text, never numeric. The values in the age column are text, not numbers! Won't be able to calculate mean of those values.

```python
df['age'].mean()
```

```python
type(np.nan)
```

```python
df['age'] = df['age'].astype(int)
```

```python
df['age'] = df['age'].astype(float)
```

```python
df.dtypes
```

```python
df['age'].mean()
```

You can cast every cell in a dataframe as the same type using df.astype(), but that's not a good idea with mixed data like this.
